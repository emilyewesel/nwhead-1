```python
columns_of_interest_nw = [
    'nwhead_only_correct_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'nwhead_full_cardio_8_0.0001_no_pretrained_resnet18_adamw_full_support_epoch21',
]
suffixes_nw = [' - balanced_acc:val:full', ' - balanced_acc:val:full:male', ' - balanced_acc:val:full:female', ' - tpr:val:full:female', ' - tpr:val:full:male', ' - tpr:val:full',]
suffixes_regular = [' - balanced_acc:val', ' - balanced_acc:val:male', ' - balanced_acc:val:female', ' - tpr:val:female', ' - tpr:val:male', ' - tpr:val']
columns_of_interest = [
    'CORAL_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'LISA_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'fchead_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'irm_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'erm_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21',
    'groupDRO_cardio_8_0.0001_no_pretrained_resnet18_adamw_correct_support_epoch21'
]

import pandas as pd


# Function to find the row with the smallest absolute difference between male and female balanced accuracy
def get_row_with_smallest_diff(column, df, suffixes, nw):
    male_col = column + ' - balanced_acc:val:male'
    female_col = column + ' - balanced_acc:val:female'
    if (nw):
      male_col = column + ' - balanced_acc:val:full:male'
      female_col = column + ' - balanced_acc:val:full:female'
    df['abs_diff'] = (df[male_col] - df[female_col]).abs()
    min_diff_row = df.loc[df['abs_diff'].idxmin()]
    return min_diff_row

# Collect results for regular columns
for column in columns_of_interest:
    min_diff_row = get_row_with_smallest_diff(column, df, suffixes_regular, False)
    step_with_min_diff = min_diff_row['Step']
    print(f"Column: {column}, Step with min diff: {step_with_min_diff}")

    for suffix in suffixes_regular:
        value = min_diff_row[column + suffix]
        print(f"Suffix: {column[:5], suffix}",  f"{value}")

# Collect results for nw columns
for column in columns_of_interest_nw:
    min_diff_row = get_row_with_smallest_diff(column, df, suffixes_nw, True)
    step_with_min_diff = min_diff_row['Step']
    print(f"Column: {column}, Step with min diff: {step_with_min_diff}")

    for suffix in suffixes_nw:
        value = min_diff_row[column + suffix]
        print(f"Suffix: {column[:11], suffix}",  f"{value}")
```
